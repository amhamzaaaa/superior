```python
height, width = img_2.shape[:2]
img_4_ = cv2.resize(img_4, (width, height))

img_4_.shape
```

```python
img_1_3 = cv2.addWeighted(img_4_, 0.6, img_2, 0.4, 0)
plt.imshow(img_1_3)
plt.axis('off')
plt.show()
```
