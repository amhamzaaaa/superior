```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

# preprocess

```python
df = pd.read_csv(r"C:\Users\123\Desktop\university\sem 3\AI lab\Assignments\task-10\data.csv")
df
```

```python
df.dropna(inplace = True)
```

```python
print(df['price'].mean())
print(df['price'].max())
```

```python
col = df.columns.tolist()
for i in col:
    if df[i].dtype == 'float64':
        df[i] = df[i].apply(lambda x: (x - df[i].min()) / (df[i].max() - df[i].min()))
```

# lable encoder

```python
a = []
for i in col:
    if df[i].dtypes == 'object':
        a.append(i)
print(a)
```

```python
LE = LabelEncoder()
for i in a:
    df[i] = LE.fit_transform(df[i])
```

```python
df['condition'] = df['condition'].apply(lambda x: (x - df[i].min()) / (df[i].max() - df[i].min()))
```

```python
df.drop(['street', 'yr_built', 'yr_renovated', 'statezip', 'country', 'condition', 'date'], axis = 1, inplace=True)
```

```python
df
```

# train-test split

```python
X = df.drop(columns='city')
Y = df['city']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state= 42)
```

# train model

```python
svc_model = SVC()
```

```python
svc_model.fit(x_train, y_train)
```

```python
y_pred = svc_model.predict(x_test)
```

```python
accu = accuracy_score(y_test, y_pred)
print(f"accuracy is: {int(accu*100)}%")
```

```python
print(classification_report(y_test, y_pred))
```
